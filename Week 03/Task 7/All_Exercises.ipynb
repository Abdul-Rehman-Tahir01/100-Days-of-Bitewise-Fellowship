# All the imports used in the exercises 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random as rd  # For exercise 01
```

### Cleaning and Pre-processing the dataset for further use in exercises

```python
# Loading dataset
titanic_dataset = pd.read_csv('Titanic.csv')

# Printing shape, head and tail of dataset
print(titanic_dataset.shape, '\n\n')
print(titanic_dataset.head().to_string(), '\n\n')
print(titanic_dataset.tail().to_string(), '\n\n')
```

```python
# Checking for missing values
print(titanic_dataset.isnull().sum())
```

```python
# Removing and replacing the null values
titanic_dataset.drop(columns='Cabin', inplace=True)
titanic_dataset.ffill(inplace=True)  # Forward fill

# Checking for missing values if remains
print(titanic_dataset.shape, '\n')
print(titanic_dataset.isnull().sum())
```

<p>-> Now, there are no missing values in the dataset</p>

### -> Exercise 1: Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
# Taking a list of values 
list_values = []

for i in range(10):
    list_values += [rd.randint(0,99)]
print(list_values)

# Plotting a simple line plot to display trend
plt.figure(figsize=(10, 5))
plt.plot(list_values)
plt.title('Simple Line Plot', fontsize=15)
plt.xlabel('Time')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```


### -> Exercise 2: Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

```python
# Selecting pclass as a feature
category_1 = titanic_dataset['Pclass'].value_counts().index
frequency_1 = titanic_dataset['Pclass'].value_counts().values

# Selecting sex as a feature
category_2 = titanic_dataset['Sex'].value_counts().index
frequency_2 = titanic_dataset['Sex'].value_counts().values

# Plotting bar plot to show the frequency distribution of Pclass and Sex
fig, (plt1, plt2) = plt.subplots(1, 2, figsize=(15, 7))

plt1.bar(category_1, frequency_1, edgecolor='black', width=0.5)
plt1.set_title('Frequency Distribution of Pclass')
plt1.set_xlabel('Categories')
plt1.set_ylabel('Frequency')

plt2.bar(category_2, frequency_2, edgecolor='black', width=0.25)
plt2.set_title('Frequency Distribution of Sex')
plt2.set_xlabel('Categories')
plt2.set_ylabel('Frequency')

plt.show()
```


### -> Exercise 3: Create a scatter plot using Matplotlib to visualize the relationship between two variables in a dataset.

```python
# Scatter plot to visualize age and fare from the dataset
plt.figure(figsize=(10, 4), facecolor='lightgray')
plt.scatter(titanic_dataset['Age'], titanic_dataset['Fare'], zorder=2)
plt.title('Age vs Fare Scatter Plot', fontsize=15)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True, zorder=1)
plt.show()
```


### -> Exercise 4: Load a dataset using Seaborn's built-in dataset functions and create a pairplot to visualize the relationships between all pairs of features.

```python
# Loading built-in iris dataset from seaborn
iris_dataset = sns.load_dataset('iris')
print(iris_dataset.head(), '\n\n')
print(iris_dataset.isnull().sum())
```

```python
# Creating a pairplot with species as hue to differentiate
sns.pairplot(iris_dataset, hue='species')
plt.title('Pairplot of Iris dataset', x=-1.25, y=4.5, fontsize=18)
plt.show()
```


### -> Exercise 5: Create a box plot using Seaborn to show the distribution of values for different categories in a dataset.

```python
# Box plot of Fare by Sex to show distribution of values of different categories
plt.figure(figsize=(10, 5))
sns.boxplot(x='Sex', y='Fare', data=titanic_dataset, color='orange')
plt.title('Box Plot of Fare by Sex')
plt.show()
```


### -> Exercise 6: Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.

```python
# Computing the correlation matrix of titanic dataset
correlation_matrix = titanic_dataset[['Age', 'Fare', 'Pclass', 'Survived']].corr()
correlation_matrix
```

```python
# Plotting heatmap to visualize correlation matrix
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds')
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```


### -> Exercise 7: Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.

```python
# Creating subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 7))

# Plotting different charts in the subplot grid
# Line plot
axes[0, 0].plot(titanic_dataset['Age'].tolist())    
axes[0, 0].set_title('Line Plot')

# Bar chart
axes[0, 1].bar(titanic_dataset['Pclass'].value_counts().index, titanic_dataset['Pclass'].value_counts().values)
axes[0, 1].set_title('Bar Chart')

# Scatter plot
axes[1, 0].scatter(titanic_dataset['Age'], titanic_dataset['Fare'])
axes[1, 0].set_title('Scatter Plot')

# Histogram Chart
axes[1, 1].hist(titanic_dataset['Age'].tolist(), bins=30, edgecolor='black')
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```


### -> Exercise 8: Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.

```python
# Customizing box plot
sns.set_palette('muted')
plt.figure(figsize=(10, 5))
sns.boxplot(x='Pclass', y='Age', data=titanic_dataset)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```


### -> Exercise 9: Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.

```python
# Creating violin plot to visualize the distribution across categories
plt.figure(figsize=(10, 4), facecolor='lightgray')
sns.violinplot(x='Sex', y='Fare', hue='Pclass', data=titanic_dataset, zorder=2)
plt.title('Violin Plot of Fare by Sex')
plt.grid(True, zorder=1)
plt.show()
```


### -> Exercise 10: Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.

```python
# Histogram with KDE plot
plt.figure(figsize=(10, 4), facecolor='lightgray')
sns.histplot(titanic_dataset['Age'], kde=True, color='blue', bins=30, zorder=2)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age distribution with KDE')
plt.grid(True, zorder=1)
plt.show()
```
