# Predicting Housing Prices using Boston Housing Dataset

### -> Importing necessary Libraries

```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### -> Loading the Boston Housing Dataset

```python
# Importing Boston Housing Dataset 
housing_data = pd.read_csv("BostonHousing.csv")
```

### -> Reading Head and Tail of dataset

```python
# Printing the head of dataset
print(housing_data.head())
```

```python
print(housing_data.tail())
```

### -> Checking for null-values and printing information

```python
#Check for missing values
print(housing_data.isnull().sum())
```

```python
print(housing_data.info())
```

```python
# Get basic statistics
print(housing_data.describe())
```

### -> Splitting the dataset

```python
#Define the feature matrix (X) and target vector (Y)
X = housing_data.drop('medv', axis=1)  # Since MEDV is the target variable
Y = housing_data['medv']

#Splitting the dataset into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# test_size = 0.2 means that 20% percent of the 100% data is reserved as test data. The remaining 80% is the training data
```

### -> Training the Linear Regression model

```python
#Creating instance of Linear Regression model
lr_model = LinearRegression()

#Training the model
lr_model.fit(X_train, Y_train)
```

```python
print(lr_model.score(X_test, Y_test))
```

### -> Printing model's coefficient and intercept

```python
#Printing model's co-efficients and intercepts
print("Co-efficients: ", lr_model.coef_)
print("Intercept: ", lr_model.intercept_)
```

### -> Predicting housing prices on test data and printing MSE

```python
#Predicting housing prices on test set
Y_pred = lr_model.predict(X_test)
```

```python
MSE = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error: ", MSE)
```

### -> Visualizing the regression line and data points

```python
#Visualize predictions vs actual values
plt.figure(figsize=(10, 6))
sns.scatterplot(x=Y_test, y=Y_pred, alpha=0.7)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual prices vs Predicted prices")

min_val = min(Y_test.min(), Y_pred.min())
max_val = max(Y_test.max(), Y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linewidth=2)
plt.show()
```
